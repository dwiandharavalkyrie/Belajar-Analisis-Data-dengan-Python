```python
#Membersihkan Data customers_df
#Menghilangkan duplicate data, jumlah duplikasi data = 6
customers_df.drop_duplicates(inplace=True)
```

```python
#memeriksa apakah duplikasi data sudah dihilangkan
print("Jumlah duplikasi: ", customers_df.duplicated().sum())
```

```python
#Menangani missing value
#menggunakan metode imputation untuk menangani missing value
#alasan = baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. 
#amputation = menggunakan nilai tertentu untuk mengganti missing value
#menggunakan nilai yang dominan sebagai pengganti missing value tersebut
#menggunakan method value_counts()untuk mengidentifikasi nilai yang dominan.
customers_df.gender.value_counts()
```

```python
#nilai yang paling dominan dalam kolom gender ialah “Prefer not to say”
#Nilai inilah yang selanjutnya akan kita gunakan sebagai pengganti missing value. 
#Proses penggantian ini dapat dilakukan menggunakan method fillna()
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
#memastikan proses amputation missing value berjalan dengan baik
customers_df.isna().sum()
```

```python
#Menangani inaccurate value
#terdapat pada nilai max yang tidak akurat pada baris age
customers_df[customers_df.age == customers_df.age.max()]
```

```python
#inaccurate value tersebut terjadi karena human error sehingga kelebihan memasukkan nilai nol. 
#mengganti age 700 menjadi 70 (yakali ada orang umur 700 T_T)
customers_df['age'] = customers_df['age'].replace(customers_df['age'].max(), 70)
```

```python
#Memeriksa apakah proses penggantian sudah selesai
customers_df.describe()
```

```python
#Membersihkan Data orders_df
#mengganti tipe data pada kolom order_date & delivery_date menjadi datetime. 
datetime_columns = ["order_date", "delivery_date"]
 
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
#memeriksa proses 
orders_df.info()
```

```python
#Membersihkan Data product_df
#terdapat 6 data yang terduplikasi pada product_df. 
#membuang data yang sama tersebut menggunakan method drop_duplicates()
product_df.drop_duplicates(inplace=True)
```
